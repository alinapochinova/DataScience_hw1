Метод **info()**— для оценки общей структуры датасета (количество записей, названия колонок, типы данных). Это позволит выявить, например, несоответствия типов данных.

**Визуальный анализ помощью countplot** — для оценки  категориальных признаков. Позволяет увидеть все уникальные значения в каждой категориальной колонке и оценить их частоту. Это необходимо, чтобы убедиться, что все значения принадлежат допустимому множеству, нет опечаток и распределение категорий выглядит правдоподобно.  
Цель: увидеть частоту каждой категории (поэтому countplot)

```python
cat_columns = dataset.select_dtypes(include=['object']).columns
n_cat = len(cat_columns)
cols = 3
rows = math.ceil(n_cat / cols)

fig_cat, axes_cat = plt.subplots(rows, cols, figsize=(cols * 10, rows * 7))
axes_cat = axes_cat.flatten()

for i, col in enumerate(cat_columns):
    ax = axes_cat[i]
    order = dataset[col].value_counts().index
    counts = dataset[col].value_counts()
    total = len(dataset)

    sns.countplot(data=dataset, x=col, ax=ax, order=order, palette="pastel", legend=False)
    ax.set_title(col, fontsize=16)
    ax.tick_params(axis='x', rotation=45)

    for p in ax.patches:
        count = int(p.get_height())
        percent = 100 * count / total
        ax.annotate(f'{count} ({percent:.1f}%)',
                    (p.get_x() + p.get_width() / 2., p.get_height()),
                    ha='center', va='bottom', fontsize=11)
    ax.set_ylim(0, max(counts.values) * 1.15)

for j in range(i + 1, len(axes_cat)):
    fig_cat.delaxes(axes_cat[j])

fig_cat.suptitle("Категориальные признаки - распределение значений", fontsize=24)
plt.tight_layout(rect=[0, 0, 1, 0.97])
plt.show()
```

Для каждого категориального признака набор уникальных значений (их количество, содержание и распределение) соответствует ожидаемому.

###3.4 Обработка категориальных признаков

####3.4.1 Применимость пункта к данным  
В данном датасете девять категориальных признаков, а большинство алгоритмов машинного обучения работают только с числовыми данными, поэтому категориальные признаки необходимо преобразовать в числовой формат. Разделим их на три типа:  
Бинарные признаки: Gender, family_history_with_overweight, FAVC, SMOKE, SCC.  
Номинальные (многоклассовые) признаки: CAEC, CALC, MTRANS.  
Целевая переменная: NObeyesdad (ординальный признак с 7 классами).
Каждый из этих типов требует своего подхода в обработке.

####3.4.2 Обоснование выбора методов

Для преобразования признаков использован **ColumnTransformer**, который позволяет применить разные методы к разным колонкам в одном объекте.  
Для бинарных признаков применен **OrdinalEncoder**. Этот метод выбран, потому что:  
1) Он преобразует категории в числа (0, 1), что идеально подходит для бинарных данных.
2) В отличие от OneHotEncoder, он не создаст лишних колонок.  

Для номинальных признаков применен **TargetEncoder**. Этот метод выбран, потому что:
1) В отличие от One-Hot Encoding, который создает много новых колонок, TargetEncoder создает одну колонку, заменяя категорию на среднее значение целевой переменной для этой категории.
2) Кодировка отражает связь категории с целевой переменной, что может улучшить качество модели.
3) У категорий нет порядка.

Для целевой переменной применен **LabelEncoder**

```python
!pip install category_encoders
```

```python
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder
from sklearn.compose import ColumnTransformer
from category_encoders import TargetEncoder
import pandas as pd
import numpy as np
```

```python
df_encoded = dataset.copy()

le = LabelEncoder()
obesity_order = [
    'Insufficient_Weight',
    'Normal_Weight',
    'Overweight_Level_I',
    'Overweight_Level_II',
    'Obesity_Type_I',
    'Obesity_Type_II',
    'Obesity_Type_III'
]
le.fit(obesity_order)
df_encoded['NObeyesdad'] = le.transform(df_encoded['NObeyesdad'])

X = df_encoded.drop('NObeyesdad', axis=1)
y = df_encoded['NObeyesdad']

binary_cols = ['Gender', 'family_history_with_overweight', 'FAVC', 'SMOKE', 'SCC']
nominal_cols = ['CAEC', 'CALC', 'MTRANS']

cat_cols_all = binary_cols + nominal_cols
num_cols = [col for col in X.columns if col not in cat_cols_all]

preprocessor = ColumnTransformer(
    transformers=[
        ('num', 'passthrough', num_cols),
        ('ordinal', OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1), binary_cols),
        ('target', TargetEncoder(), nominal_cols)
    ],
    verbose_feature_names_out=False
)

X_processed = preprocessor.fit_transform(X, y)

feature_names = num_cols + binary_cols + nominal_cols
X_processed_df = pd.DataFrame(X_processed, columns=feature_names)
```

```python
X_processed_df.head()
```

```python
print("Бинарные признаки после OrdinalEncoder:")
for col in binary_cols:
    print(f"{col}: {X_processed_df[col].unique()}")

print("Номинальные признаки после TargetEncoder:")
for col in nominal_cols:
    print(f"{col}: уникальных значений: {X_processed_df[col].nunique()}; диапазон: [{X_processed_df[col].min():.3f}, {X_processed_df[col].max():.3f}]")
```

**Визуальный анализ с помощью boxplot** - для идентификации выбросов в числовых признаках. Это позволит четко увидеть точки, являющиеся аномальными значениями.

```python
original_numeric_cols = ['Age', 'Height', 'Weight', 'FCVC', 'NCP', 'CH2O', 'FAF', 'TUE']

n_num = len(original_numeric_cols)
cols = 3
rows = math.ceil(n_num / cols)

fig_num, axes_num = plt.subplots(rows, cols, figsize=(cols * 5, rows * 4))
axes_num = axes_num.flatten()

for i, col in enumerate(original_numeric_cols):
    ax = axes_num[i]
    sns.boxplot(data=X_processed_df, x=col, ax=ax, width=0.3)
    ax.set_title(col, fontsize=12)
    ax.set_xlabel("")

for j in range(i + 1, len(axes_num)):
    fig_num.delaxes(axes_num[j])

fig_num.suptitle("Анализ выбросов в исходных числовых признаках", fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.97])
plt.show()
```

Метод **3 сигмы** — в качестве базового фильтра для оценки выбросов. Значения, отклоняющиеся от среднего более чем на 3 стандартных отклонения, будут считаться потенциальными выбросами.

```python
original_numeric_cols = ['Age', 'Height', 'Weight', 'FCVC', 'NCP', 'CH2O', 'FAF', 'TUE']
data_numeric = X_processed_df[original_numeric_cols].copy()

def outlier_std(data, col, threshold=3):
    mean = data[col].mean()
    std = data[col].std()
    up_bound = mean + threshold * std
    low_bound = mean - threshold * std
    anomalies = pd.concat([data[col] > up_bound, data[col] < low_bound], axis=1).sum(axis=1).astype(bool)
    return anomalies

std_outliers = pd.Series(data=False, index=data_numeric.index)

for col in original_numeric_cols:
    anomalies = outlier_std(data_numeric, col)
    std_outliers[anomalies[anomalies].index] = True

print(f"Найдено {std_outliers.sum()} выбросов")
print(f"Доля выбросов: {std_outliers.sum()/len(data_numeric)*100:.2f}%")

labeled_data = data_numeric.copy()
labeled_data['is_outlier'] = std_outliers.astype(int)

sns.pairplot(data=labeled_data, vars=original_numeric_cols,
             hue='is_outlier', hue_order=[1, 0],
             palette=['red', 'blue'],
             diag_kind='kde',
             plot_kws={'alpha': 0.6, 's': 20},
             diag_kws={'alpha': 0.6})

plt.suptitle("Выбросы по методу 3 сигм", y=1.02, fontsize=16)
plt.tight_layout()
plt.show()
```

Метод **межквартильного отклонения (IQR)** - не зависит от предположения о нормальности распределения данных, поэтому устойчив к ассиметричным распределениям. Необходимо применить оба метода для более объективной оценки.IQR использует медиану и квартили.

```python
original_numeric_cols = ['Age', 'Height', 'Weight', 'FCVC', 'NCP', 'CH2O', 'FAF', 'TUE']
data_numeric = X_processed_df[original_numeric_cols].copy()

def outlier_iqr(data, col, threshold=6):
    IQR = data[col].quantile(0.75) - data[col].quantile(0.25)
    up_bound = data[col].quantile(0.75) + (IQR * threshold)
    low_bound = data[col].quantile(0.25) - (IQR * threshold)
    anomalies = pd.concat([data[col] > up_bound, data[col] < low_bound], axis=1).sum(axis=1).astype(bool)
    return anomalies

iqr_outliers = pd.Series(data=False, index=data_numeric.index)

for col in original_numeric_cols:
    anomalies = outlier_iqr(data_numeric, col, threshold=3)
    iqr_outliers[anomalies[anomalies].index] = True

print(f"Найдено {iqr_outliers.sum()} выбросов")
print(f"Доля выбросов: {iqr_outliers.sum()/len(data_numeric)*100:.2f}%")

labeled_data = data_numeric.copy()
labeled_data['is_outlier'] = iqr_outliers.astype(int)

sns.pairplot(data=labeled_data, vars=original_numeric_cols,
             hue='is_outlier', hue_order=[1, 0],
             palette=['red', 'blue'],
             diag_kind='kde',
             plot_kws={'alpha': 0.6, 's': 20},
             diag_kws={'alpha': 0.6})

plt.suptitle("Выбросы по методу межквартильного размаха (IQR)", y=1.02, fontsize=16)
plt.tight_layout()
plt.show()
```

Результаты по методу 3 сигм и метожу межквартильного отклонения дали слишком разные результаты, поэтому проведем дополнительный анализ.

```python
# Анализ распределения данных
for col in original_numeric_cols:
    print(f"{col}")
    print(f"Среднее: {data_numeric[col].mean():.2f}")
    print(f"Медиана: {data_numeric[col].median():.2f}")
    print(f"Станд. отклонение: {data_numeric[col].std():.2f}")
    print(f"Q1: {data_numeric[col].quantile(0.25):.2f}")
    print(f"Q3: {data_numeric[col].quantile(0.75):.2f}")
    print(f"IQR: {data_numeric[col].quantile(0.75) - data_numeric[col].quantile(0.25):.2f}")
    print(f"Границы 3 сигм: [{data_numeric[col].mean()-3*data_numeric[col].std():.2f}, {data_numeric[col].mean()+3*data_numeric[col].std():.2f}]")
    print(f"Границы IQR*3: [{data_numeric[col].quantile(0.25)-3*(data_numeric[col].quantile(0.75)-data_numeric[col].quantile(0.25)):.2f}, {data_numeric[col].quantile(0.75)+3*(data_numeric[col].quantile(0.75)-data_numeric[col].quantile(0.25)):.2f}]")

    anomalies_3sigma = outlier_std(data_numeric, col)
    anomalies_iqr = outlier_iqr(data_numeric, col, threshold=3)
    print(f"Выбросы 3 сигм: {anomalies_3sigma.sum()}")
    print(f"Выбросы IQR*3: {anomalies_iqr.sum()}")
```

Можем увидеть, что из-за распределения признака NCP (дискретное, грубо говоря 3 значения) IQR очень низкий. 334 выброса, которые считает IQR, - это на самом деле не выбросы, просто значения признака.

Для выявления шумовых признаков будет использован комплексный подход, включающий методы фильтрации и встроенные методы:  
1) **Корреляционный анализ** — позволит выявить сильно коррелирующие признаки, которые дублируют информацию.

4) Анализ важности признаков с помощью **случайного леса** — для оценки важности каждого признака. В отличие от линейного F-теста, случайный лес способен улавливать нелинейные зависимости и взаимодействия между признаками.  
5) Анализ **зависимости качества модели от количества признаков** — позволит оценить, как меняется качество предсказаний при последовательном добавлении признаков, отсортированных по важности. Это даст возможность выбрать оптимальное количество признаков.
